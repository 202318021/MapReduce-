```python
from pyspark.sql import SparkSession
from pyspark import SparkContext, SparkConf
```

```python
config = SparkConf()
config.setMaster("local").setAppName("MapReduce")
print(f' Master URL: {config.get("spark.master")}')
print(f' App Name: {config.get("spark.app.name")}')
```

```python
spk_cntxt = SparkContext.getOrCreate()
spk_cntxt
```

```python
spark = SparkSession.builder.appName("MapReduce").master("local").getOrCreate() 
```

```python
 spark
```

```python
path = 'dbfs:/FileStore/shared_uploads/kandarpmalkan3@gmail.com/Blog1-1.txt'
rdd = spark.sparkContext.textFile(path) # RDD created
```

```python
lines = spk_cntxt.textFile(path)
```

```python
rdd.collect()
```

```python
split_rdd = rdd.flatMap(lambda x: x.split(' '))
words = split_rdd.collect()
words
```

```python
word_freq = split_rdd.map(lambda x: (x,1))
word_freq.collect()
```

```python
count_rdd = word_freq.reduceByKey(lambda a,b: a+b)
count_rdd.collect()
```

```python
rdd_sorted = count_rdd.map(lambda x: (x[1],x[0])).sortByKey()
rdd_sorted
```

```python
first_ten = count_rdd.take(10)
for f in first_ten:
    print(f" : {f[0]} Value: {f[1]} ")
```

```python
total_words = rdd_sorted.reduce(lambda a,b: (a[0]+b[0],a[1]))
print(f"Total Words : {total_words[0]} ")
```

```python
data = rdd.collect()
data
```

```python
import re
pattern = r'[\n\s!@#$%^&*\(\)\{\}\[\]~`/\?<>,\.]+'
words = re.split(pattern, str(data))
words
```

```python
word_count = list(map(lambda x: (x, 1), words))
word_count
```

```python
word_counts = {}
for word in words:
    word = word.lower()
    word_counts[word] = word_counts.get(word, 0) + 1

print(f'Total number of words:  {sum(word_counts.values())}')
```

```python
print('\n\nList of Words: ')
print('\n\nWord \t\tFrequency ')
for key, values in word_counts.items():
    if key == '':   continue
    print(f' {key}: \t {values} ')
```
